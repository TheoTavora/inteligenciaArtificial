```python
import numpy as np
import pandas as pd
from sklearn.feature_selection import mutual_info_classif
from sklearn.preprocessing import LabelEncoder
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report

# Definindo o número de funcionários
n_funcionarios = 1000

# Gerando características aleatórias
np.random.seed(42)
idade = np.random.randint(20, 60, n_funcionarios)
anos_empresa = np.random.randint(0, 20, n_funcionarios)
satisfacao = np.random.uniform(0, 1, n_funcionarios)
horas_mensais = np.random.randint(100, 250, n_funcionarios)
aumento_salario = np.random.choice([0, 1], n_funcionarios, p=[0.7, 0.3])
curso_capacitacao = np.random.choice([0, 1], n_funcionarios, p=[0.6, 0.4])
projetos = np.random.randint(1, 10, n_funcionarios)
departamento = np.random.choice(['Recursos Humanos', 'Vendas', 'TI', 'Produção'], n_funcionarios)

# Gerando o rótulo (target) de forma balanceada
rescisao = np.random.choice(['Rescisão Voluntária', 'Rescisão Involuntária', 'Continua na empresa'],
                            n_funcionarios, p=[0.2, 0.1, 0.7])

# Criando o DataFrame
df = pd.DataFrame({
    'Idade': idade,
    'Anos na Empresa': anos_empresa,
    'Satisfação no Trabalho': satisfacao,
    'Horas Mensais Trabalhadas': horas_mensais,
    'Recebeu Aumento de Salário': aumento_salario,
    'Curso de Capacitação': curso_capacitacao,
    'Número de Projetos': projetos,
    'Departamento': departamento,
    'Resultado': rescisao
})

# Visualizando os primeiros registros do dataset
print(df.head())

# Salvando o dataset em um arquivo CSV para uso posterior
df.to_csv('funcionarios.csv', index=False)
```

```python
contagem_resultado = df['Resultado'].value_counts()

# Exibir a contagem
print(contagem_resultado)
```

```python
df.drop(['Departamento'], axis=1, inplace=True)
```

```python
df.rename(columns={'Resultado': 'Class'}, inplace=True)

df['Class'].replace({True: 0, False: 1}, inplace=True)
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import mutual_info_classif

data = df
Y = data['Class']
X = data.drop(columns=['Class'])
np.random.seed(7)

X_array = X.to_numpy()
y_series = pd.Series(Y)

information_gain = mutual_info_classif(X_array, y_series)

for feature_name, info_gain in zip(X.columns, information_gain):
    print(f"Informação {feature_name}: {info_gain}")
```

```python
# VAMOS JOGAR AS ETIQUETAS PARA Y
Y = df['Class'].values
print(Y)

# X SERÁ A NOSSA MATRIZ COM AS FEATURES
X = df[['Idade', 'Horas Mensais Trabalhadas', 'Anos na Empresa']].values
print(X)
```

```python
X_treino, X_teste, Y_treino, Y_teste = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
model = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=500)
model.fit(X_treino, Y_treino)

Y_pred = model.predict(X_teste)

accuracy = accuracy_score(Y_teste, Y_pred)
print("Acurácia do modelo:", accuracy)

print(classification_report(Y_teste, Y_pred))
```
